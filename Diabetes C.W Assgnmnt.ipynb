```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score, recall_score, f1_score
```

```python
#dataset loaded
df = pd.read_csv(r"C:\Users\Rizwan\IDS\Excel\diabetess.csv")
```

```python
df
```

```python
df.head()
```

```python
#zero values replaced with NaN
non_zero = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
df[non_zero] = df[non_zero].replace(0, np.nan)
```

```python
df
```

```python
#filling null values with the mean of the data
df.fillna(df.mean(), inplace=True)
df
```

```python
# Check for any remaining missing values
print(df.isnull().sum())
```

```python
X = df.drop('Outcome', axis=1)
y = df['Outcome']
X
```

```python
y
```

```python
#standardizing the features
scaler = StandardScaler()
scaled_X= scaler.fit_transform(X)
scaled_X
```

```python
# Visualizing the distribution 
sns.countplot(x='Outcome', data=df)
plt.title('Distribution of Diabetes')
plt.show()
```

```python
df.hist(bins=10, figsize=(10, 10))
plt.tight_layout()
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, test_size=0.2, random_state=42)

```

```python
# Creating and training the Logistic Regression model
model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
predict_y = model.predict(X_test)
predict_y
```

```python
# Calculate evaluation metrics
accuracy = accuracy_score(y_test,predict_y)
precision = precision_score(y_test,predict_y)
recall = recall_score(y_test,predict_y)
f1 = f1_score(y_test,predict_y)
```

```python
print(f'Accuracy: {accuracy:.2f}')
print(f'Precision: {precision:.2f}')
print(f'Recall: {recall:.2f}')
print(f'F1 Score: {f1:.2f}')
```

```python
# Confusion Matrix
conf_matrix = confusion_matrix(y_test,predict_y)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
# Classification Report
class_report = classification_report(y_test, predict_y)
print(class_report)
```
